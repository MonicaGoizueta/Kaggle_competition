```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
import pickle

# para calcular las métricas

from sklearn.model_selection import cross_validate
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor 
from sklearn import tree

from sklearn.linear_model import LinearRegression
from sklearn.metrics import make_scorer
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder  
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
plt.rcParams["figure.figsize"] = (10,8)
```

```python
df_tt = pd.read_csv("../data/preproc.csv", index_col = 0)
df_tt.head(2)
```

```python
X = df_tt.drop('price', axis =1)
y = df_tt['price']
```

# Ajustamos el modelo usando GridSearch

```python
param = {"max_depth": [4, 5, 6, 7],
        "min_samples_split": [10, 50, 100],
        "max_features": [1,2,3,4,5,6]}

bosque = GridSearchCV(
            estimator=RandomForestRegressor(),
            param_grid= param,
            cv=10,
            verbose=3,
            return_train_score = True,
            scoring="neg_mean_squared_error")
```

```python
%time
bosque.fit(X, y)
```

```python
bos = bosque.best_estimator_
bos
```

# Predicción

Nos traemos el CSV de test limpio

```python
X_test = pd.read_csv("../data/test_limpio_.csv", index_col = 0)
X_test.head(2)
```

```python
y_pred_test_rf = bos.predict(X_test)
y_pred_train_rf = bos.predict(X)
```

```python
y_pred_test_rf
```

# Validación del Modelo

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_pred_train_rf))
print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred_train_rf))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_pred_train_rf)))
print("R2:",  metrics.r2_score(y, y_pred_train_rf))
```

```python
def metricas(y, y_pred_test_rf, y_pred_train_rf, tipo_modelo):
    
    
    resultados = {'Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_pred_train_rf))}
    df = pd.DataFrame(resultados)
    df["modelo"] = tipo_modelo
    return df
```

```python
rf_results = metricas(y, y_pred_test_rf, y_pred_train_rf, "Random Forest")
rf_results
```
